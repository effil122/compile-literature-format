```python
import docx
import pandas as pd
import re
```

```python
doc = docx.Document("Mami.docx")
```

```python
# print("段落の個数：", len(doc.paragraphs))
# print("表の個数：", len(doc.tables))
# print("図（行内）の個数：", len(doc.inline_shapes))
```

```python
refs = []
for para in doc.paragraphs:
    if para.text[:1] == "[":
        refs.append(para.text)
```

```python
refs_fix = []

for ii,ref in enumerate(refs):
    reff=ref.replace('['+str(ii+1)+"]","")
    ref_list = reff.replace('.', ' ').replace(',', '').split()
    ref_list2 = reff.replace('.', '').replace('\t', '').split(',')
#     print(reff)
    ref_fix=""
    count=0
    for i,s in enumerate(ref_list):
        if len(s) == 1 and len(ref_list[i-1]) != 1:
            count += 1
            if count > 3: #4人以上用 et al
                if count == 4:
                    ref_fix = ref_fix[:-2]
                    ref_fix += " et al  "
                    continue
                else:
                    continue
            if len(ref_list[i+1]) != 1: #人数追記
                ref_fix += ref_list[i+1]+" "+s+", " 
            elif len(ref_list[i+2]) != 1: #ミドルネーム用
                ref_fix += ref_list[i+2]+" "+s+ref_list[i+1]+", " 
            else:
                ref_fix += ref_list[i+3]+" "+s+ref_list[i+1]+ref_list[i+2]+", " 
        else:
            if len(s) != 1 and len(ref_list[i+1]) != 1 and ref_list[i+1] != "and":
                break
    ref_fix = ref_fix[:-2]
    for k,ref2 in enumerate(ref_list2):
        if ref_list[i+1] in ref2:
            title = ref2
            break
    try:
        ref_fix += " "+"("+ref_list[-1].replace('(',"").replace(')',"")+")"+ title+". "+ref_list2[k+1].replace(' ',"")+":"+ref_list2[k+2].replace(' ',"")
        refs_fix.append(ref_fix)
    except Exception as e:
        refs_fix.append("Error! "+str(e)+reff)
        print(e)
    print(ref_fix)
# print(refs_fix)
```

```python
refs_fix.sort()
doc2 = docx.Document()
for ref_fix in refs_fix:
    print(ref_fix)
    doc2.add_paragraph(str(ref_fix))
doc2.save("fix.docx")
```
